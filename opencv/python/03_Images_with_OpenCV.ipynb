### 3.3 Drawing on Images

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
blank_img = np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
plt.imshow(blank_img)
```

```python
blank_img = cv2.rectangle(blank_img,pt2=(384,0),pt1=(510,150),color=(0,255,0),thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
# cv2.circle()
# cv2.line()
# thickness = -1 -> fill
```

```python
cv2.putText(blank_img,
            text='Hello',
            org=(10,500),
            fontFace=cv2.FONT_HERSHEY_SIMPLEX,
            fontScale=4,
            color=(255,255,255),
            thickness=3,
            lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

```python
# filled polygon
vertices = np.array([[100,300],[200,200],[400,300],[200,400]], dtype=np.int32) # shape = (4,2)
pts = vertices.reshape((-1,1,2)) # this step is required so that each channel has a point
cv2.fillPoly(blank_img,[pts],color=(0,0,255)) # pts passed as a list
plt.imshow(blank_img)
```
