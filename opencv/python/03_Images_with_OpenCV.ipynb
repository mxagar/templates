### 3.1 Open & Handle Images

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
img = cv2.imread('../data/00-puppy.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

```python
img_gray = cv2.imread('../data/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img_gray, cmap='gray')
```

```python
fix_img.shape
```

```python
img_resized = cv2.resize(img_gray, (1000, 400))
```

```python
plt.imshow(img_resized)
```

```python
width_ratio = 0.5 # 50% of the original
height_ratio = 0.2 # 20% of the original
new_img = cv2.resize(fix_img, (0,0), fix_img, width_ratio, height_ratio) 
```

```python
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img, 0) # 0: upside down, 1: left-right
```

```python
new_img = cv2.cvtColor(new_img,cv2.COLOR_RGB2BGR) # undo conversion
cv2.imwrite('totally_new.jpg', new_img) # format is especified in file ending
```

```python
fig = plt.figure(figsize=(10,8)) # 10x8 inches space for canvas
ax = fig.add_subplot(1,1,1)
ax.imshow(fix_img)
```
