## 6.3 Canny Edge Detection

```python
# Canny algorithm: a widely used edge detection method: : multi-stage process that performs
# - Noise reduction (Gaussian blurring)
# - Gradient computation: Sobel X and Y
# - Non-maximum supression: some gradient values lead to pixels to be filtered out
# - Double-thresholding: low and high
# - Hysteresis thresholding: it is determined if a pixel is edge-like or not, weak edges are supressed
#     - below low, removed
#     - above high taken
#     - between low and high: taken iff connected to edges with high
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../../data/sammy_face.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(img, threshold1=127, threshold2=127)
```

```python
plt.imshow(edges)
```

```python
# Nice heuristic for choosing the thresholds
median = np.median(img)
lower = int(max(0,0.7*median)) # 30% below median
upper = int(max(255,1.3*median)) # 30% above median
```

```python
# Always manually apply a Gaussian blur!
blurred_image = cv2.blur(img, ksize=(4,4))
```

```python
# We can tune the heuristacally derived thresholds!
edges = cv2.Canny(blurred_image, threshold1=lower, threshold2=upper-50)
```

```python
plt.imshow(edges)
```
