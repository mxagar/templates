### Keyboard Control

```python
# First we need to give focus to a window, for instance by clicking
```

```python
import pyautogui
```

```python
pyautogui.position()
```

```python
# Click on window, then type string with 0.2 secs interval between characters
pyautogui.click(624,728); pyautogui.typewrite('Hello world!', interval = 0.2)
```

```python
# All special keys we can pass
# ESC, volume up/down
pyautogui.KEYBOARD_KEYS
```

```python
# Type a -> ab -> xab -> yxab
pyautogui.typewrite(['a', 'b', 'left', 'left', 'x', 'y'], interval=1)
```

```python
# Press a single key
pyautogui.press('f1')
```

```python
# Press several keys simultaneously: Command+S
pyautogui.hotkey('command', 's')
```

### Screenshots and Image Recognition

```python
# On Linux machines, scrot needs to be installed to use screenshots
# sudo apt-get install scrot
```

```python
import pyautogui
```

```python
# Get a PIL image object of the screen: screenshot
pyautogui.screenshot()
```

```python
# If we pass a path, image is saved there (use absolute paths)
pyautogui.screenshot('/Users/mxagar/Desktop/screenshot.png')
```

```python
# We can request to recognize/locate an image on screen!
# But: it lasta around 1 sec, pixels must be almost identical
# We need to have image files of the elements we'd like to find and pass them
# Top-left corner + H & W are returned
```

```python
save_position = pyautogui.locateOnScreen('/Users/mxagar/Desktop/save.png')
```

```python
save_position
```

```python
# We can now move there and click
pyautogui.moveTo (save_position.left + save_position.width*0.5, save_position.top + save_position.height*0.5, duration = 1)
pyautogui.click()
```
